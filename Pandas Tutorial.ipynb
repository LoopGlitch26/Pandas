## header = 'infer' and names = None

By default, when we read any dataset, the parameter header is set to be 'infer'. Infer stands for inferring the header names from the dataset itself.

If no names are passed as column names in the read_csv() function, header becomes equal to 'infer' and Pandas automatically pass in the first row elements as column names by default. In this case, names are set to None

If names are passed explicitly, header becomes equal to none and those names are used as Column Names.

Let us see this with the help of examples

```python
draft2 = pd.read_csv('pokemon_data.csv')
```

Here, the names of the columns are not explicitly mentioned, so header is equal to infer and column names are the first row values of the dataset

```python
colnames = list(draft1.columns)
colnames.append('New Name')


draft3 = pd.read_csv('pokemon_data.csv',names=colnames)
```

```python
draft3
```

Here, the column names are explicitly mentioned and passed in the read_csv() function so this time header becomes equal to None and these names are used as column names. 

Notice that the usual column names occupied the first row this time

## nrows parameter

```python
draft1.head(30)
```

```python
draftx = pd.read_csv('pokemon_data.csv',nrows = 30)

draftx
```

Just by first look, we can see that both head() function and nrows parameter returns the same result but nrows has another use which we will see now.

* When we use head() function and pass 30 as a parameter, the "entire" dataframe gets readed and first 30 rows are displayed.

* But when we use the parameter nrows = 30, only the first 30 rows are read not the entire data and gets displayed.

Uses of nrows:-

While dealing with a large dataset consisting of GBs of data, if we try to use head(), the computational speed reduces. But if we use nrows and just display the necessary rows, the task can easily be done without wasting RAM.

## chunksize parameter

The chunksize parameter is similar to the nrows wherein the entire data is not usually read. 

But the difference lies in the size of the data in bytes. By setting chunksize as x bytes, the data is read till it reaches the x bytes and is stopped if it immediately exceeds the specified size.

These are also useful when dealing with larger datasets where the size is known.

Lets see an example

```python
chunk = pd.read_csv('pokemon_data.csv',chunksize=100)

#When we use chunksize, a text file is returned so we iterate through it to display data

for i in chunk:
    print(i)
```

## Writing Data to a Text Format

Till now, we have seen how to read data from a dataset and display it as a dataframe in pandas.

But in real life, once we finish up editing and modifying the dataset, we would sometimes need to save it as a file in our local storage to use it another time.

This task can be accomplished in pandas by using to_csv() function.

```python
#Now, lets see how to write the data into a csv format.

c1 = pd.read_csv('pokemon_data.csv',nrows = 100)

c1
```

```python
c1.to_csv('Compressed_Pokemon_Data.csv')

#Now, a new csv file appears with the name Compressed_Pokemon_Data which has only 100 entries
```

If we want to save the dataset into another directory, we can specify the complete path inplace of only the name and get the job done, just like with reading data from another directory


Let us look at a few more scenarios and try to solve them.

1. Sometimes, we would like to store the data by using a different seperator and not just commas.

2. In the case of Online Machine Learning Competitions, usually we have to upload the predictions data without any header or index for the algorithm to scan and display our score.

For the scenario one, we can use the parameter sep = '@' to get our job done. This is similar to the syntax of sep in read_csv() function as well.

Now, lets do this operation and look at the resulting dataset

```python
c1.to_csv('Compressed_Pokemon_Data.csv',sep='@')

dfa = pd.read_csv('Compressed_Pokemon_Data.csv')

dfa
```

As we can see, the data is modified and @ is the new sepearator now.

In scenario 2, we don't want the data to have an index or column names while writing. We can accomplish this task by using two parameters called 'header' and 'index'.

The default value of 'header' and 'index' are set as True so we need to explicitly change them inorder to not get column names and index.

Now, lets code this

```python
c1 = pd.read_csv('pokemon_data.csv',nrows = 100)

c1.to_csv('Compressed_Data.csv',header=False,index=False)

c11 = pd.read_csv('Compressed_Data.csv')

c11.head()
```

Now, we can clearly see that both the column names and the index are missing when we read the dataset.

Now, there might be cases where we want to view the data first before writing it to a file name to see if everything worked out well, rather than writing it and then going to the folder to see.

This can be done by using a default library in python called 'sys' 

```python
import sys

draft1.to_csv(sys.stdout,na_rep='NO VALUE')

#na_rep replaces the nan values with a default value in this case NO VALUE. sys.stdout displayes the text in clipboard
```

By using the function 'sys.stdout', we can print the data in the clipboard and check it before writing it to a file

# JSON DATA

JSON stands for "Java Script Object Notation".

JSON is also another popular type of formats to represent data so it is important to know how to read and manipulate them by using Pandas.

Typically, JSON data looks like a complex connections of data within data in the form of dictionaries and lists.

Let us look at an example below

```python
obj = '''
{
	"id": "0001",
	"type": "donut",
	"name": "Cake",
	"ppu": 0.55,
	"batters":
		{
			"batter":
				[
					{ "id": "1001", "type": "Regular" },
					{ "id": "1002", "type": "Chocolate" },
					{ "id": "1003", "type": "Blueberry" },
					{ "id": "1004", "type": "Devil's Food" }
				]
		},
	"topping":
		[
			{ "id": "5001", "type": "None" },
			{ "id": "5002", "type": "Glazed" },
			{ "id": "5005", "type": "Sugar" },
			{ "id": "5007", "type": "Powdered Sugar" },
			{ "id": "5006", "type": "Chocolate with Sprinkles" },
			{ "id": "5003", "type": "Chocolate" },
			{ "id": "5004", "type": "Maple" }
		]
}
'''
type(obj)
```

The data inside the ''' is the JSON data and in this case, we are intialising it as a string to work with Pandas.

The default way is to save the data in .json format and use a default function 'read_json()' to read the json directly. But for now, let us look at how we can work with JSON data in strings.

```python
import json
result = json.loads(obj)
result
```

Here, we used the library 'json' inorder to display the json data.

The .loads() method in json takes in the string of json data as an input and returns the dictionary as an output

```python
type(result)
```

In order to convert the dictionary data into string format, we can use the .dumps() method and pass in the dictionary data as an input. This method is shown in the below cell

```python
backtostr = json.dumps(result)
type(backtostr)
```

Now that we have successfully converted the JSON data from string to dictionary, we can use a portion of the data or the entire data to create a datadrame in Pandas.

Additionally we can specify only those column names which we need to further filter the data

```python
topping = pd.DataFrame(result['topping'],columns=['id','type'])
topping
```

```python
result['topping']
```

We can index even a dictionary inside a dictionary by specifying the appropriate index slicing as shown below.

In this example, we created a dataframe of batter dictionary which is available inside batters dictionary which is a part of the result dictionary

```python
pd.DataFrame(result['batters']['batter'],columns=['id','type'])
```

# Web APIs

An API stands for "Application Programming Interface" 

It acts as a connection between two homogeneous or heterogeneous systems or databases.

APIS work on the principle of "Request and Response". Let us understand it in a Leymann way.

When we try to withdraw money from an ATM, the ATM first sends a request to the bank server to check if the required amount of money is available in the account. This is called 'Request'.

Once it recieves a confirmation that money is available, it sends a response to the ATM to start giving out the money. This is called a 'Response'.


An API acts as a bridge between these two systems ATM and Bank Server and integrates the functionalities of both

```python
import requests
url= 'https://api.github.com/repos/pandas-dev/pandas/issues'
resp = requests.get(url)
resp

#request and response way.
```

We use a default library called 'requests' available in python to handle responses and requests.

In the above code, we are trying to specify the url and requesting the data from the url and storing it in an object called resp.

The output resp is given as <Response [200]>. Here 200 indicates a successful response to the request

```python
data = resp.json()
data
```

Now that we obtained the response successfully, we proceed to read the data.

Since the data is avaialable in JSON, we use the .json() function to read the data.

Now, that we have read the data, we can perform modifications and create dataframes from it just like we did it in the JSON section
